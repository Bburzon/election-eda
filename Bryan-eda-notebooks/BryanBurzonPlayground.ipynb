# Data Cleaning 
1. filter to onlypresidential data 
2. Extract state from location column 
    - assumption: that values with 2 states are actually targeting the state that is the swing state 

```python
# pres_data = np.char.replace(pres_data.location, "Iowa", "IA")

pres_data.location = pres_data.location.str.replace("Iowa", "IA")

pres_data.location = pres_data.location.str.replace("Ohio", "OH")

pres_data.location.value_counts()
```

3. Extract state from location and add new State Column

```python
pres_data["state"] = pres_data.location.str[-2:]
pres_data.state.value_counts()
```
